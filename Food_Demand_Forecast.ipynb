```python
cat_columns = df.select_dtypes(['object']).columns
#convert all categorical variables to numeric
df[cat_columns] = df[cat_columns].apply(lambda x: pd.factorize(x)[0])
```

```python
df
```

# Applying normalization on checkout_price and base_price columns

```python
df['checkout_price'] = (df['checkout_price'] - df['checkout_price'].mean()) / df['checkout_price'].std() 
df['base_price'] = (df['base_price'] - df['base_price'].mean()) / df['base_price'].std() 
```

##Now, let's do the same thing on test set as well!

```python
df_test['checkout_price'] = (df_test['checkout_price'] - df_test['checkout_price'].mean()) / df_test['checkout_price'].std() 
df_test['base_price'] = (df_test['base_price'] - df_test['base_price'].mean()) / df_test['base_price'].std() 
```

```python
df_test
```

```python
new_df_test = pd.merge(df_test, df_meal)
df_test = pd.merge(new_df_test,df_center)
df_test = df_test[['id','week','center_id','city_code','region_code','center_type','op_area','meal_id','category','cuisine','checkout_price','base_price','emailer_for_promotion','homepage_featured']]
```

```python
cat_columns = df_test.select_dtypes(['object']).columns
#convert all categorical variables to numeric
df_test[cat_columns] = df_test[cat_columns].apply(lambda x: pd.factorize(x)[0])
```

#It's time for Model training!

```python
y = df['num_orders']
df.drop('num_orders', axis=1, inplace=True)
x = df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

#XGBoost Regression Model 

```python
from sklearn.metrics import mean_squared_error as MSE
import xgboost as xgb
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
xg_reg.fit(X_train,y_train)
preds = xg_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

#Gradient Boost Regression Model

```python
from sklearn.metrics import mean_squared_log_error
from sklearn.ensemble import GradientBoostingRegressor
GB = GradientBoostingRegressor()
GB.fit(X_train, y_train)
y_pred = GB.predict(X_test)
y_pred[y_pred<0] = 0
RMSEL=mean_squared_log_error(y_test, y_pred)
RMSEL=np.sqrt(RMSEL)
RMSEL
```

#Random Forest Regression Model

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF.fit(X_train, y_train)
y_pred = RF.predict(X_test)
y_pred[y_pred<0] = 0
RMSEL=mean_squared_log_error(y_test, y_pred)
RMSEL=np.sqrt(RMSEL)
RMSEL
```

#Predicting the number of orders of each individual IDs for weeks 145-155.

```python
pred_test_data= RF.predict(df_test)
pred_test_data[pred_test_data<0] = 0
submit_RF = pd.DataFrame({
    'id' :df_test['id'],
    'num_orders' : pred_test_data
})
```
